## Connecting to the Database
Import the necessary packages and connect to the database **data.sqlite**.

```python
#Your code here
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite')
c = conn.cursor()
```

```python
def show(dataframe):
    dataframe.columns = [x[0] for x in c.description]
    return dataframe
```

## Display the names of all the employees in Boston.
Hint: join the employees and customers tables.

```python
#Your code here
c.execute("""select firstName,
                    lastName
                    from employees e
                    join offices o
                    using(officeCode)
                    where o.city = "Boston";""")
df = pd.DataFrame(c.fetchall())
show(df)
```

## Do any offices have no employees?
Hint: Combine the employees and offices tables and use a group by.

```python
c.execute("""select * from offices;""")
df = pd.DataFrame(c.fetchall())
show(df)
```

```python
#Your code here
c.execute("""select city,
                    count(*)
                    from offices o
                    left join employees e
                    using (officeCode)
                    group by 1;""")
df = pd.DataFrame(c.fetchall())
show(df)
```

## Write 3 Questions of your own and answer them

* **what are the top 10 cities that spent the most in their orders?**

```python
# Your code here
c.execute("""select city,
                    amount as total_amount_spent
                    from payments p
                    join customers c
                    using (customerNumber)
                    group by 2
                    limit 10;""")
df = pd.DataFrame(c.fetchall())
show(df)
```

* **which employee has the most customers they are responsible for?**

```python
c.execute("""select firstName, 
                    lastName,
                    count(employeeNumber) as no_clients
                    from employees e
                    join customers c
                    on e.employeeNumber = c.salesRepEmployeeNumber
                    group by e.employeeNumber
                    order by count(employeeNumber) desc
                    limit 1;""")
df = pd.DataFrame(c.fetchall())
show(df)
```

```python
c.execute("""select customerName,
                    firstName,
                    lastName
                    from customers c
                    join employees e
                    on e.employeeNumber = c.salesRepEmployeeNumber
                    where firstName = 'Pamela';""")
df = pd.DataFrame(c.fetchall())
show(df).head(10)
```

* **how many orders do we have per country?**

```python
c.execute("""select country,
                    count(orderNumber) as no_orders
                    from customers c
                    join orders o
                    using(customerNumber)
                    group by 1
                    order by 2 desc;
                    """)
df = pd.DataFrame(c.fetchall())
show(df)
```
